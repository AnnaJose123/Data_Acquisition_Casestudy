```python
q1=data['area'].quantile(0.25)
q3=data['area'].quantile(0.75)
iqr=q3-q1
lower_bound=q1-1.5*iqr
upper_bound=q3-1.5*iqr
outlier=data[(data['area']<lower_bound)] | data[(data['area']>upper_bound)]
outlier
```

```python
data.groupby('stories')['price'].agg(['mean','sum'])
```

```python
price_by_airconditioning = data.groupby('airconditioning')['price'].mean()
print(price_by_airconditioning)
```

```python
mean_price = data['price'].mean()
median_price = data['price'].median()
price_75th_percentile = data['price'].quantile(0.75)
print("Mean price:", mean_price)
print("Median price:", median_price)
print("75th percentile price:", price_75th_percentile)
```

```python
moderate_furnished_houses = data[(data['price'] >= median_price - 100000) &
                                (data['price'] <= median_price + 100000) &
                                (data['bedrooms'] >= 3) &
                                (data['bathrooms'] >= 2) &
                                (data['furnishingstatus'] == 1)]
print(moderate_furnished_houses)
```

```python
filtered_houses = data[(data['furnishingstatus'] == 'furnished') &
                        (data['airconditioning'] == 'yes') &
                        (data['mainroad'] == 'yes') &
                        (data['guestroom']=='yes') &
                        (data['basement']=='yes') ]\
                        .sort_values('price')

filtered_houses
```

In this analysis, we can easily find the house corresponding to price, having all the facilities like furnished AC guestroom, etc.
